```python
import pandas as pd

# 원본 엑셀 파일 경로
file_path = 'Vehicle Count Simon Bolivar Avenue.xlsx'

# Excel 파일 객체 로드
xls = pd.ExcelFile(file_path)

# 결과 저장용 딕셔너리
processed_sheets = {}

# 모든 시트 순회
for sheet_name in xls.sheet_names:
    df = pd.read_excel(xls, sheet_name=sheet_name)

    if df.shape[0] < 4:
        continue  # 너무 적은 데이터는 건너뜀

    # 방향 이름과 센서 ID 추출
    dir_names = df.iloc[1]
    sensor_ids = df.iloc[2]

    # 새로운 컬럼명 생성
    new_columns = []
    for dir_name, sensor_id in zip(dir_names, sensor_ids):
        if pd.notna(dir_name):
            new_columns.append(str(dir_name).strip())
        else:
            new_columns.append(str(sensor_id).strip())

    # 중복된 이름 뒤쪽 5개는 " Divided4" 붙이기
    # 일단 5개의 열은 무조건 뒤쪽 5개라고 가정
    if len(new_columns) >= 5:
        for i in range(-5, 0):
                new_columns[i] = new_columns[i] + " Divided4"

    # 데이터만 추출
    data = df.iloc[3:].copy()
    data.columns = new_columns
    data = data.reset_index(drop=True)

    # 'FECHA' → 'DAY', 'Horario' → 'Time' 으로 컬럼 이름 정리
    data.rename(columns={'FECHA': 'DAY', 'Horario': 'Time'}, inplace=True)

    # 'DAY' 빈 셀 채우기
    if 'DAY' in data.columns:
        data['DAY'] = data['DAY'].fillna(method='ffill')

    # 결과에 저장
    processed_sheets[sheet_name] = data

# 모든 시트를 하나의 파일로 저장
output_file = 'Processed_Vehicle_Count_Simon_Bolivar_Avenue.xlsx'
with pd.ExcelWriter(output_file, engine='openpyxl') as writer:
    for sheet_name, sheet_data in processed_sheets.items():
        sheet_data.to_excel(writer, sheet_name=sheet_name, index=False)

print(f"전처리된 엑셀 파일이 '{output_file}'에 저장되었습니다.")
```
